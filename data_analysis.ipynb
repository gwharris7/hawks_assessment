```python
# import required packages
import pandas as pd
import matplotlib.pyplot as plt
# load raw data
df_raw = pd.read_csv('data/nbaplayersdraft.csv')
# preview data
df_raw.head()
```

```python
# get summary statistics
df_raw.describe()
```

```python
# based on NBA franchise history, change team abbreviations to match those of their present-day counterparts
df_clean = df_raw.replace(['NJN', 'CHA', 'CHH', 'NOH', 'NOK', 'SEA', 'VAN', 'WSB'], 
                          ['BRK', 'CHO', 'CHO', 'NOP', 'NOP', 'OKC', 'MEM', 'WAS'])
```

```python
# filter by players who went to Duke and were drafted in or before 2000
duke_2000 = df_clean[(df_clean['college'] == 'Duke') & (df_clean['year'] <= 2000)]
duke_2000.head()
```

```python
# count appearances for each NBA team
duke_2000_teams = duke_2000.groupby(['team'])['player'].count().reset_index(name='count')
# sort by most appearances
duke_2000_sorted = duke_2000_teams.sort_values(['count'], ascending=False)
print(duke_2000_sorted)
# Dallas, Minnesota, and Phoenix each drafted 2 players from Duke in or before the 2000 draft
```

```python
# filter by players whose first name begins with D and who were drafted in an even year
d_even = df_clean[(df_clean['player'].str.startswith('D')) & (df_clean['year'] % 2 == 0)]
d_even.tail()
```

```python
# count appearances for each NBA team
d_even_teams = d_even.groupby(['team'])['player'].count().reset_index(name='count')
# sort by most appearances
d_even_sorted = d_even_teams.sort_values(['count'], ascending=False)
print(d_even_sorted)
# Oklahoma City (including the Seattle SuperSonics) drafted 8 players whose first name begins with D in even year drafts
```

```python
picks = df_clean[['year', 'overall_pick', 'team']]
# observe that the data does not account for trades involving picks, which would have changed the original
# lottery-assigned pick order: here, I use each team's first pick from each year (their lowest, hence minimum, pick
# number from that year) as a proxy for their first round pick slot
picks_pivoted = picks.pivot_table(index='team', columns='year', values='overall_pick', aggfunc = min)
# filter for only first-round picks: here, I assume a 30-team league (despite earlier years having only 28 or 29 teams)
first_round = picks_pivoted[picks_pivoted <= 30]
# create sample graph showing first round pick history for Atlanta Hawks
first_round
hawks = first_round.loc['ATL']
plt.plot(hawks)
plt.title('Atlanta Hawks First Round Draft Picks 1989-2021')
# save graph as PNG
plt.savefig('figures/hawks_first_round.png')
plt.show()
```

```python
# construct data table with correlations
corrs = dict()
for team, picks in first_round.iterrows():
    # get the first round picks for each team
    picks = picks.dropna()
    # get the first round picks for the subsequent year and compute correlation using standard correlation coefficient
    picks_next = picks.shift(-1)
    corrs[team] = [(picks.corr(picks_next, method='pearson'))]
    # get the difference between the two years' picks
    picks_diff = picks_next - picks
    # get the mean difference
    picks_diff_mean = picks_diff.mean()
    # print the mean difference
    corrs[team].append(picks_diff_mean)
picks = pd.DataFrame.from_dict(corrs, orient='index', columns=['correlation', 'mean yr-to-yr diff.'])
print(picks)
# save data table as CSV
picks.to_csv('figures/pick_slots_yr_to_yr.csv')
print(f'Mean results across all teams:\n{picks.mean()}')
```
